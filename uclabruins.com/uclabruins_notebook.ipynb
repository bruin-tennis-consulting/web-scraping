### Load Packages

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

### Notebook Start

```python
url = "https://uclabruins.com/api/v2/Stats/boxscore/33233"
# url = "https://uclabruins.com/sports/mens-tennis/stats/2023-24/texas-am/boxscore/33233"
```

### Check Status Code
- 200 = success
- 404 = failed

```python
page = requests.get(url)
page.status_code
```

```python
match_data = BeautifulSoup(page.content, 'html.parser')
match_data = page.json()
```
