```python
stats_link = 'https://colleges.wearecollegetennis.com/Competitions/ITA/Tournaments/draws/D1CB8036-87C4-4CBD-B968-929D4500BD0E'
```

```python
session = AsyncHTMLSession()
response = await session.get(stats_link)
if response.ok:
    await response.html.arender()
    stats_soup = BeautifulSoup(response.html.html, 'html.parser')
    print('Success!')
```
